```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

sleep_db = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
sleep_db.info()
```

```python
plt.figure(figsize = (25,20))
plt.subplot(2,2,1)
plt.gca().set_title('Variable Occupation')
sns.countplot(x = 'Occupation', palette = 'Set1', data = sleep_db)

plt.subplot(2,2,2)
plt.gca().set_title('Gender')
sns.countplot(x = 'Gender', hue='Sleep Disorder', palette = 'Set3', data = sleep_db)

plt.subplot(2,2,3)
plt.gca().set_title('BMI')
sns.countplot(x = 'BMI Category', hue='Sleep Disorder', palette = 'Set2', data = sleep_db)
```

### First, we need to check the 'objects' and convert them to numeric ones.
Here, we have gender, occupation, bmi, and blood pressure

```python
print("Unique values of gender:", sleep_db['Gender'].unique())
print("Unique values of occupation:", sleep_db['Occupation'].unique())
print("Unique values of BMI:", sleep_db['BMI Category'].unique())
print("Unique values of B-Pressure:", sleep_db['Blood Pressure'].unique())
```

```python
#Handling the blood pressure
sleep_db_temp = sleep_db.copy()
new_columns = sleep_db_temp['Blood Pressure'].str.split('/', expand = True)
sleep_db_temp['B-Pressure_high'], sleep_db_temp['B-Pressure_low'] = new_columns[0], new_columns[1]
sleep_db_temp = sleep_db_temp.drop('Blood Pressure', axis = 1)
sleep_db_temp['B-Pressure_high'] = sleep_db_temp['B-Pressure_high'].astype(float)
sleep_db_temp['B-Pressure_low'] = sleep_db_temp['B-Pressure_low'].astype(float)
sleep_db_temp.info()
```

```python
sleep_db_temp = sleep_db_temp.drop('Person ID', axis = 1)
sleep_db_temp.hist(bins = 60, figsize = (20,15))
```

```python
sleep_db_temp['BMI_Category'] = sleep_db_temp['BMI Category']
sleep_db_temp['BMI_Category'] = sleep_db_temp['BMI_Category'].str.strip().str.lower()
sleep_db_temp['BMI_Category'] = sleep_db_temp['BMI_Category'].replace(['normal weight'], ['normal'])
sleep_db_temp = sleep_db_temp.drop(['BMI Category'], axis = 1)
#They looked like same

sleep_db_hot = pd.get_dummies(sleep_db_temp[['Gender', 'Occupation', 'BMI_Category']], drop_first = False)
sleep_db_temp = pd.concat([sleep_db_hot, sleep_db_temp], axis = 1)
sleep_db_temp.to_csv('sleep_db_reorganized.csv')
sleep_predict = sleep_db_temp['Sleep Disorder']
sleep_db_temp = sleep_db_temp.drop(['Gender', 'Occupation', 'BMI_Category', 'Sleep Disorder'], axis = 1)
sleep_db_temp.info()
```

```python
correlation = sleep_db_temp.corr()
plt.figure(figsize = (18,18))
sns.heatmap(correlation, annot = True)
```

There ara some minor correlated elements, but can we say that they have visible relation to each other? For example, do the high and low bp levels dependent to each other? Let's leave it for now.

Update: When i check some information about tension, the normal range always has the bigges ratio (120/80 = 1.5). When it leaves from the healthy zone, the ratio decreases (130/90 = 1.44, 180/120 = 1.28). But these ratios are not stable. So, we can't use a simple ratio.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scaler = StandardScaler()
sleep_db_temp = sleep_db_temp.values
sleep_db_values = scaler.fit_transform(sleep_db_temp)

x_Train, x_Test, y_Train, y_Test = train_test_split(sleep_db_values, sleep_predict, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

models_and_params = [
    {
        'name': 'Random Forest Classifier',
        'classf': RandomForestClassifier(),
        'params': {
            'classf__n_estimators': [100, 200, 300,400],
            'classf__max_depth': [None, 5, 10,15]
        }
    },
    {
        'name': 'Logistic Regression',
        'classf': LogisticRegression(),
        'params': {
            'classf__solver': ['liblinear', 'lbfgs'],
            'classf__C': [0.01,0.1, 1, 10],
            'classf__max_iter': [1000]
        }
    },
    {
        'name': 'K Nearest Neighbors',
        'classf': KNeighborsClassifier(),
        'params': {
            'classf__n_neighbors': [3, 5, 7,9]
        }
    },
    {
        'name': 'Gradient Boosting Classifier',
        'classf': GradientBoostingClassifier(),
        'params': {
            'classf__n_estimators': [100, 200, 300,400],
            'classf__learning_rate': [0.01, 0.1, 1]
        }
    },
    {
        'name': 'Decision Tree Classifier',
        'classf': DecisionTreeClassifier(),
        'params': {
            'classf__max_depth': [None, 5, 10,15]
        }
    }
]
accuracy_scores = []
best_estimators = []
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from yellowbrick.classifier import ConfusionMatrix

for model in models_and_params:
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('classf', model['classf'])
    ])
    grid_search = GridSearchCV(pipeline, model['params'], cv=5)
    grid_search.fit(x_Train, y_Train)
    best_model = grid_search.best_estimator_
    best_estimators.append(best_model)

    #Get accuracy score
    y_pred = best_model.predict(x_Test)
    decision_report = classification_report(y_Test, y_pred)
    accuracy = accuracy_score(y_Test, y_pred)
    accuracy_scores.append(accuracy)
    #Get report
    print(model['name'])
    print(decision_report)
```

```python
#plot the accuracy diagram
model_names = [model['name'] for model in models_and_params]
plt.figure(figsize=(15, 5))
sns.barplot(x=model_names, y=accuracy_scores)
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy of Different Models')
plt.savefig("Accuracy_diagram.png")
plt.show()

#Apply confusion matrix
for model in models_and_params:
    matrix = ConfusionMatrix(model['classf'], classes=['Insomnia', 'None', 'Sleep Apnea'])
    matrix.fit(x_Train, y_Train)
    matrix.score(x_Test, y_Test)
    plt.savefig(f"{model['name']}_confusion_matrix.png")
    matrix.show()
    plt.clf()  #Clear the current figure's content
```

All the algorithms performs better when predicting non-patient people. Logistic Regression is the best so far with 92%.
